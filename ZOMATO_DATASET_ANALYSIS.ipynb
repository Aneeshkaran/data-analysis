```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df=pd.merge(df,df_country,on='Country Code',how='left')
df.head()
```

```python
country_names=df.Country.value_counts().index
country_values=df.Country.value_counts().values
```

```python
plt.pie(country_values[:3],labels=country_names[0:3],autopct='%1.2f%%')
```

```python
df.columns
```

```python
rating=df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
rating
```

```python
sns.barplot(x='Aggregate rating',y='Rating count',data=rating,hue='Rating color',palette=['blue','red','orange','yellow','lightgreen','darkgreen'])
```

```python
sns.countplot(x="Rating color",data=rating,palette=['blue','red','orange','yellow','green','green'])
```
